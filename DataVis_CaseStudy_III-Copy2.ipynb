# Time Series And Animation #

Series of data points that are indexed in time order. 
The independent variable is the time.

Some examples : Stock Market Data, ECG, Weather Data, COVID Data...

```python
#Load plotly 
import plotly.express as px
import plotly.graph_objs as go
import numpy as np
import datetime
import pandas as pd
```

```python
df = px.data.stocks()
df
```

```python
px.line(df, x='date', y='NFLX')
```

```python
f2 = go.Figure()
f2.add_trace(go.Scatter(x=df['date'], y=df['NFLX'],mode='lines'))
f2.show()
```

```python
f2.update_xaxes(range=['2019-01-01', '2019-04-01'])
```

```python
f2 = go.Figure()
f2.add_trace(go.Scatter(x=df['date'], y=df['NFLX'],mode='lines'))
f2.update_xaxes(rangeslider_visible=True)
```

```python
f3 = go.Figure()
f3.add_trace(go.Scatter(x=df['date'], y=df['GOOG'],mode='lines',name='GOOG'))
f3.update_xaxes(rangeslider_visible=True)
f3.update_xaxes(rangeselector=dict(
        buttons=list([
            dict(count=5, label="5D", step="day", stepmode="backward"),
            dict(count=1, label="1M", step="month", stepmode="backward"),
            dict(count=3, label="3M", step="month", stepmode="todate"),
            dict(count=6, label="6M", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1Y", step="year", stepmode="backward"),
            dict(label="MAX", step="all")
        ])))
```

```python
f3 = go.Figure()
f3.add_trace(go.Scatter(x=df['date'], y=df['GOOG'],mode='lines',name='GOOG'))
f3.update_xaxes(rangeslider_visible=True)
f3.update_layout(hovermode='x unified', template='simple_white')
f3.update_traces(fill='tonexty',line_color='blue')
```
